```python
import pandas as pd
```

```python
df = pd.read_excel('seguros.xlsx')
pd.set_option('display.max_rows', df.shape[0]+1)
```

## Detalle de facturas

```python
df.groupby(['RAZON_SOCIAL_EMISOR', 'RUC_EMISOR','FECHA_EMISION', 'SERIE_COMPROBANTE']).sum()
```

## Resumen por año

```python
data = df.groupby(['RAZON_SOCIAL_EMISOR', df['FECHA_EMISION'].dt.year]).sum()
data
```

```python
pd.pivot_table(data, index='FECHA_EMISION', columns='RAZON_SOCIAL_EMISOR', values='IMPORTE_TOTAL')
```

```python
data.agg(TOTAL=('IMPORTE_TOTAL','sum'))
```
